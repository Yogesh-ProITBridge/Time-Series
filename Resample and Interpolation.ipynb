```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
shampoo_sales = pd.read_csv("/content/sales.csv", index_col=0,parse_dates=True)
```

```python
df = shampoo_sales.copy()
```

```python
plt.figure(figsize=(13,5))
df.plot()
plt.title('Month-wise shampoo sales',size=15)
plt.show()
```

```python
df
```

```python
day_df = df.resample("D").mean()
day_df
```

```python
day_df = day_df.interpolate(method='linear')
day_df
```

```python
plt.figure(figsize=(16,5))
plt.plot(day_df['Sales'])
plt.show()
```

## **Downsampling** - **Quarterly**

```python
qt_df = day_df.resample("Q").mean()
qt_df
```

```python
plt.figure(figsize=(16,5))
plt.plot(qt_df['Sales'])
plt.show()
```

```python
Year_df = qt_df.resample('Y').mean()
Year_df
```

```python
plt.figure(figsize = (10,5))
plt.plot(Year_df['Sales'])
plt.show()
```
